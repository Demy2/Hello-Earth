```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import pyplot as plt
```

```python
sample_data=pd.read_excel('Book12.xlsx',sheet_name='Sheet1')
```

```python
sample_data
```

```python
plt.bar(sample_data['Aspect'],sample_data['Positive'])
plt.xlabel('Aspects',fontsize=15)
plt.ylabel('Positive', fontsize=15)
plt.xticks(rotation='vertical')
plt.title('Aspect Base Analysis \n with respect to Trump Tweets',fontsize=16)
plt.savefig('myplot1.png')
```

```python
plt.show()
```

```python
plt.bar(sample_data['Aspect'],sample_data['Negative'])
plt.xlabel('Aspects',fontsize=15)
plt.ylabel('Negative', fontsize=15)
plt.xticks(rotation='vertical')
plt.title('Aspect Base Analysis \n with respect to Trump Tweets',fontsize=16)
plt.savefig('myplot2.png',color='red')
```

```python
plt.show()
```
